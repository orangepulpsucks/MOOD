```python
# CV2 using DLIB
# This library contains emotions: angry, disgust,fear, happy, sad, surprise, neutral

# Case 1 - Correct

# Imports
import cv2
from fer import FER
import matplotlib.pyplot as plt

# Load Images
image_path = '/Users/y/Desktop/Coding/CS330/Final Project/AI_Expression_Recognition/happy.jpg'
image = cv2.imread(image_path)

# Initialize FER Detector
detector = FER()

# Detect Faces in Image
faces = detector.detect_emotions(image)

# Process each detected face
for face in faces:
    emotion = max(face['emotions'], key=face['emotions'].get)
    (x, y, w, h) = face['box']

    # Draw a rectangle around the face
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Display the emotion label
    cv2.putText(image, f"Emotion: {emotion}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

# Display the results using matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
