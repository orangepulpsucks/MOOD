```python
# GOOGLE CLOUD VISION API

# Case 1 - Correct

# Imports
from google.cloud import vision_v1
import io
import cv2
import matplotlib.pyplot as plt

# api key: 
# service accnt id: vision-api-422

# Set Google Cloud Vision API Key
api_key_path = '/Users/y/Desktop/Coding/CS330/Final Project/aiexpressionrecognition-95def551db84.json'
client = vision_v1.ImageAnnotatorClient.from_service_account_file(api_key_path)

# Read Image
image_path = 'happy.jpg'
with io.open(image_path, 'rb') as image_file:
    content = image_file.read()

# Create Vision API Image Object
image = vision_v1.Image(content=content)

# Perform Emotion Analysis
response = client.face_detection(image=image)
faces = response.face_annotations

# Display Image
img = cv2.imread(image_path)
plt.imshow(img[:, :, :: -1])
plt.show()

# Check if faces are detected
if faces:
    for i, face in enumerate(faces):
        print(f"Emotions for face {i + 1}:", 
              f"Joy: {face.joy_likelihood},",
              f"Sorrow: {face.sorrow_likelihood},",
              f"Anger: {face.anger_likelihood},",
              f"Surprise: {face.surprise_likelihood}",
              f"Blurred: {face.blurred_likelihood}",
              f"Headwear: {face.headwear_likelihood}")

    # Example: Get the dominant emotion for the first face
    dominant_emotion = max(['joy', 'sorrow', 'anger', 'surprise', 'blurred', 'headwear'], key=lambda x: getattr(faces[0], f'{x}_likelihood'))
    print()
    print(f"The emotion this person is feeling is {dominant_emotion}.")
else:
    print("No face detected in the image.")
```
